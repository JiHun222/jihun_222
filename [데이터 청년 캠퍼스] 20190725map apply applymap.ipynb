```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# 남자를 1 여자를 2로 하고 싶다면
# map 사용. map은 series에서만 사용 가능
tips.sex.map({'Male':1, 'Female':2})
# 이렇게 딕셔너리 써도 되고 함수 써도 된다
```

```python
tips.sex.map(lambda x : 1 if x == 'Male' else 2 ) # series에 있는 데이터가 하나하나씩 들어간다
```

```python
tips.sex.apply(lambda x : 1 if x == 'Male' else 2)
# map은 series 아니면 DataFrame에서만
```

```python
tips.apply(lambda x : x['total_bill']+x['tip'], axis = 1)
# apply에는 axis 개념이 숨어있다 series 할 때는 map과 동일
# dataframe에서는 조금 더 복잡하게 쓸 수 있다
```

```python
tips[['total_bill','tip']].apply(lambda x : sum(x))
```

```python
# apply, applymap만 알면 모든 기능 사용 가능
```

```python
# for 안 쓰는 애
# 1. 이터,제너
# 2. comprerehension
# 3. 재귀함수
# 4. map/filter/reduce
```

```python
tips[['total_bill','tip']].applymap(lambda x : x+1) #map과는 비슷하지만 얘는 각각
```

```python
tips['ratio'] = tips.apply(lambda x:x['tip']/x['total_bill'], axis = 1) # 비율 나온 값 추가
```
