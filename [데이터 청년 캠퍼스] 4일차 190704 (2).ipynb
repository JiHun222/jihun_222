# map reduce 복습

```python
from functools import reduce
```

```python
map(lambda x:x+1, [1,2,3,4])
```

```python
reduce(lambda x,y:x+y, [1,2,3,4]) # 2개 이상인 애가 1개로 축약되는 것
                                  # 2개에 값이 들어간다
```

```python
reduce(lambda x,y:x+y, [1,2,3,4], 5) # 처음에 (5,1)이 들어간다
                                     # 초기값이 없으면 (1,2) emfdjrksek
```

```python
reduce(lambda x,y:x+[y], [1,2,3,4], []) # x에 5들어가고, y에 1 들어가면 type error
                                       # 5대신 비어있는 list [ ] 를 집어넣자
                                       # 결과 1이 elemental을 가지는 list가 된다 
```

```python
a = [str.upper, str.lower]
```

```python
import numpy as np
```

```python
np.all
```

```python
all_pred = lambda item, *tests: all(p(item) for p in tests)
any_pred = lambda item, *tests: any(p(item) for p in tests)
```

# partial

```python
from functools import partial
from operator import add
```

```python
add3 = partial(add,3)
```

```python
add3(7) # add를 3으로 대치시킨다
        # closure를 쉽게 만들어주는 기법
        # return을 함수로 했기 때문에 partail도 higher-order function
```

# closure

```python
def make_add(x):
    def inner(y):
        return x+y
    return inner
```

```python
make_add(3)(4)
```
